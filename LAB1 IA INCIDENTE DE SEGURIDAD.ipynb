```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
n_samples = 1000
equipos_afectados = np.random.normal(50, 10, n_samples).reshape(-1, 1)
costo = 1000 + 10 * equipos_afectados + np.random.normal(0, 50, n_samples).reshape(-1, 1)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(equipos_afectados, costo, c='b', label='Datos')
plt.xlabel('Equipos afectados')
plt.ylabel('Costo del incidente')
plt.title('Dataset: Equipos afectados vs. Costo del incidente')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(equipos_afectados, costo, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(X_train, y_train, "b.", label="Datos de entrenamiento")
plt.plot(X_test, y_test, "g.", label="Datos de prueba")

plt.plot(X_train, y_train_pred, "r-", label="Regresión lineal")

x_new = np.array([[60]])  # Ejemplo de equipos afectados
coste = model.predict(x_new)
plt.plot(x_new, coste, "rx", markersize=10, label="Predicción para 60 equipos")

plt.xlabel("Equipos afectados")
plt.ylabel("Costo del incidente")
plt.title('Incidentes de Seguridad en una Empresa')
plt.legend()
plt.show()
```
